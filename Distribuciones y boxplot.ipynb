```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sns
```

```python
%matplotlib inline
```

```python
datos = np.random.randn(100)
sns.distplot(datos)
```

```python
sns.distplot(datos,rug=False,hist=False) #solo la curva
```

```python
#distribuciones argumentos
arg_curva = {'color':'black','label':'Curva'}
arg_histograma = {'color':'green','label':'Histograma'}
sns.distplot(datos, bins=25,kde_kws=arg_curva, hist_kws=arg_histograma)
```

```python
#seaborn+pandas
```

```python
serie = pd.Series(datos)
serie
```

```python
sns.distplot(serie,bins=25,color="green")
```
