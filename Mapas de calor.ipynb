```python
#mapas de calor (heatmap)  http://seaborn.pydata.org/generated/seaborn.heatmap.html
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#dataset de vuelos
```

```python
vuelos = sns.load_dataset('flights')
vuelos.head()
```

```python
#creamos una matriz para el mapa de calor
```

```python
vuelos = vuelos.pivot('month','year','passengers')
```

```python
vuelos
```

```python
sns.heatmap(vuelos)
```

```python
#ver valores
```

```python
sns.heatmap(vuelos,annot=True,fmt='d')
```

```python
#recoger un valor de la matriz
```

```python
valor_central = vuelos.loc['May'][1956]
valor_central
```

```python
#cambiar valores de colores
```

```python
sns.heatmap(vuelos,center=valor_central,annot=True,fmt='d')
```

```python
#moviendo la barra de colores
```

```python
sns.heatmap(vuelos,center=valor_central,annot=True,fmt='d',cbar_kws={'orientation':'horizontal'})
```
