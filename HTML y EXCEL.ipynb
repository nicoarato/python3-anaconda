```python
import pandas as pd
```

```python
url='https://es.wikipedia.org/wiki/Copa_Mundial_de_F%C3%BAtbol'
```

```python
dataframe = pd.io.html.read_html(url)
```

```python
dataframe
```

```python
data_futbol = dataframe[0]
data_futbol
```

```python
columnas = data_futbol.loc[0]
columnas = dict(columnas)
data_futbol = data_futbol.rename(columns=columnas)
data_futbol
```

```python
data_futbol = data_futbol.drop(0)
data_futbol
```
