```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#utilizamos datos por default
```

```python
propinas = sns.load_dataset('tips')
propinas.head()
```

```python
sns.lmplot('total_bill','tip',propinas)
```

```python
#agregar marcadores y colores
```

```python
sns.lmplot('total_bill','tip',propinas,scatter_kws={'marker':'o','color':'green'},
                                        line_kws={'color':'blue'})
```

```python
#quitando la linea
```

```python
sns.lmplot('total_bill','tip',propinas,fit_reg=False)
```

```python
propinas['porcentaje'] = 100*propinas['tip']/propinas['total_bill']
propinas.head()
```

```python
sns.lmplot('size','porcentaje',propinas) 
```

```python
sns.lmplot('total_bill','porcentaje',propinas,hue='sex',markers=['x','o']) #propinas por sexo
```

```python
sns.lmplot('total_bill','porcentaje',propinas,hue='day') #propinas por dias
```
